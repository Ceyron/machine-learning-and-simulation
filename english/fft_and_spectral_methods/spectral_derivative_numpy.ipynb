# Spectral derivative in NumPy

$$ f'(x) = \text{ifft}(i k \; \text{fft}(f(x))) $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 1
N = 100
```

```python
# IMPORTANT: The right boundary point is not included in the mesh!
mesh = np.linspace(0, L, N, endpoint=False)
```

```python
f = lambda x: np.sin(4 * np.pi * x)
f_prime = lambda x: 4 * np.pi * np.cos(4 * np.pi * x)
```

```python
f_h = f(mesh)
f_prime_h = f_prime(mesh)
```

```python
plt.plot(mesh, f_h, label="original")
plt.plot(mesh, f_prime_h, label="analytical derivative")
plt.grid()
plt.legend()
```

```python
k = np.fft.fftfreq(N, d=L/(N * 2 * np.pi))
```

```python
f_prime_fft_h = np.fft.ifft(1j * k * np.fft.fft(f_h)).real
```

```python
plt.plot(mesh, f_h, label="original")
plt.plot(mesh, f_prime_h, label="analytical derivative")
plt.plot(mesh, f_prime_fft_h, label="spectral derivative")
plt.grid()
plt.legend()
```
